```python
import ROOT
from pyroot_cms_scripts import CMS_style, CMS_text
```

```python
ROOT.TGaxis().SetMaxDigits(4)
CMS_style.SetTitleYOffset(1.0)
CMS_style.cd()
ROOT.gROOT.ForceStyle()

data_file = ROOT.TFile.Open("hists/double_muon_data_2017full.root")
mc_file = ROOT.TFile.Open("hists/dy_amcatnlo_20_mc_2017full.root")

title = ";#Delta#phi (Z, jet);p^{T}_{jet} / p^{T}_{Z}"

data_dphi_zj_ptj_z = data_file.Get("data_dphi_zj_ptj_z")
data_dphi_zj_ptj_z.SetTitle(title)
prompt_dphi_zj_ptj_z = mc_file.Get("prompt_dphi_zj_ptj_z")
prompt_dphi_zj_ptj_z.SetTitle(title)
pileup_dphi_zj_ptj_z = mc_file.Get("pileup_dphi_zj_ptj_z")
pileup_dphi_zj_ptj_z.SetTitle(title)
```

```python
plot = """
canvas = ROOT.TCanvas("", "", 700, 600)
canvas.SetRightMargin(0.13)
data_dphi_zj_ptj_z.Draw("colz")
canvas.SetLogz(1)
canvas.Draw()
canvas.Print("plots/data_dphi_zj_ptj_z.pdf")
"""
```
